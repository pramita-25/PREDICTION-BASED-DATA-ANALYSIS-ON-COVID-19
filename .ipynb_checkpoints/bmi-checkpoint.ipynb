```python
#pandas for play with data
import pandas as pd

#numpy use for some array
import numpy as np

#seaborn use for making statistical graphics. Seaborn is a data visualization library
import seaborn as sns
```

```python
data = pd.read_csv("bmi.csv")
```

```python
#Scikit-learn or sklearn is a free software machine learning library
#Support vector machines (SVMs) are a set of supervised learning methods used for classification, regression and outliers detection
from sklearn import svm
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data.head()
```

```python
gender = LabelEncoder()
data["Gender"] = gender.fit_transform(data['Gender'])
```

```python
data.head()
```

```python
bins = (-1,0,1,2,3,4,5)
health = ['malnourished', 'underweight', 'fit', 'slightly overweight', 'overweight', 'extremly overweight']
data["Index"] = pd.cut(data['Index'], bins=bins, labels=health)
```

```python
sns.relplot(x="Index", y="Height", hue="Gender", kind='line', data = data, height=5, aspect=3)
```

```python
x = data.drop('Index', axis=1)
y = data['Index']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 0)
```

```python
s = StandardScaler()
x_train = s.fit_transform(x_train)
x_test = s.transform(x_test)
```

```python
clf = svm.SVC()
clf.fit(x_train, y_train)
pred = clf.predict(x_test)
```

```python
print(classification_report(y_test, pred))
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(accuracy_score(y_test, pred))
```
