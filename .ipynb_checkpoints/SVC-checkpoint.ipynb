```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#pandas #numpy #matplotlib #seaborn
```

```python
dataset = pd.read_csv('covid19_patient_data.csv')

dataset.head()

#no transformation required. All are numerical data i.e., encoded
```

```python
corrmat = dataset.corr()
f, ax = plt.subplots(figsize=(7, 7))
sns.heatmap(corrmat, vmax=.8, square=True, annot=True, fmt='.2f')
plt.show()

#heatmap
#Correlation Matrix
```

```python
dataset.Target.unique()
x=dataset[['Fever','ThroatInfections','BreathingIssue']]

y=dataset['Target']

#defining input variables x and y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)

#training and testing data spliting
```

```python
from sklearn.svm import SVC

classifier = SVC()
classifier
#classifier (to get parameters)
#max_depth criterion max_features n_estimators
```

```python
classifier.fit(x_train, y_train)

#model can be ready to classify using fit
```

```python

score = classifier.score(x_test, y_test)
print(score)

#to obtain score value for evaluation of classifier (accuracy)
```
